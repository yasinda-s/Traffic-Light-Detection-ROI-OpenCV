```python
import cv2

roi_x = 600
roi_y = 480
roi_width = 120
roi_height = 150

lower_green = (50, 120, 10)
upper_green = (83, 255, 37)

video_capture = cv2.VideoCapture('green_light.webm')

if not video_capture.isOpened():
    print("Error: Could not open video file.")
    exit()

while True:
    #ret is bool to see if frame was obtained or not, frame is the actual frame from the video
    ret, frame = video_capture.read()
    if not ret:
        break

    #frame can be adjusted like a singular image using cv2 operations
    frame = cv2.resize(frame, (1024, 768))
    # cv2.imshow('Original', frame)
    frame_with_ROI = frame.copy()
    #to frame_with_ROI, add the ROI
    cv2.rectangle(frame_with_ROI, (roi_x, roi_y), (roi_x + roi_width, roi_y + roi_height), (255, 255, 255), 2)
    cv2.putText(frame_with_ROI, "ROI", (roi_x, roi_y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    cv2.imshow('ROI', frame_with_ROI)

    # extract the ROI
    roi = frame[roi_y:roi_y + roi_height, roi_x:roi_x + roi_width]
    #mask to get only green pixels
    mask = cv2.inRange(roi, lower_green, ideal_green)
    #count number of pixels that survived the mask (green range)
    num_green_pixels = cv2.countNonZero(mask)

    if num_green_pixels > 0:
        print("Green light detected, drive now!")
        # pixel_value = roi[10, 10] 
        # print(pixel_value) #to gauge correct lower and upper green
    else:
        print("Not green light, stop driving!")

    #33 to make approximately 30 FPS
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
